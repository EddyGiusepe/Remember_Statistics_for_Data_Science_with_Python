## 1. Medidas de posição

### Média ou média aritmética

Exemplo:

* Média de notas anuais
* Média de vendas
* Média de idades


image.png

```python
# Por exemplo: vamos a calcular a média das notas (de algumas matérias) de um aluno

notas = [8, 4, 9.5, 7.33, 10, 3.6, 10]

def media_notas(vetor):
  return sum(vetor)/len(vetor)


print("A média das notas deste aluno é:", media_notas(notas))  
print("")
print("A média pode ser calculada, também, assim: ", np.average(notas))
print("")
print("Ou assim: ", np.mean(notas))

# OBS: A média é sensível a valores extremos
```

A seguir vamos analisar como a MÉDIA é alterada para certos valores.  

```python
valores = {'lista1': [1, 5, 7, 9, 100]} # valores --> 5, 7, 9, 1, 100

valores = pd.DataFrame(valores)
valores.head()
```

```python
media = valores.lista1.mean()

plt.hist(valores.lista1)
plt.axvline(x = media, color = 'r', linestyle='--')

print("A média, é:", media)

# OBS: Outliers, valores muito distantes da distribuição, eles afeitam muito a média.
# Para um valor no extremo a média estará como contrapeso para o outro extremo
```

### Mediana

* Valor que divide o conjunto de Dados em duas partes iguais

* Para calcular a mediana precisamos ordenar o conjunto de Dados 

OBS:

    * Se o tamanho do conjunto de Dados for ÍMPAR, então a mediana será o valor central
    * Se o tamanho do conjunto de Dados for PAR, então a mediana será a média aritmética dos dois valores centrais.

\\
Exemplo:

```python
dados_1 = [10, 13, 11, 10, 87]
dados_2 = [23, 31, 17, 18]


print("A mediana de dados_1, é:", np.median(dados_1))
print("")

print("A mediana de dados_2, é:", np.median(dados_2))
```

### Moda

* Valor com maior frequência em um conjunto de Dados

* Podemos ter as seguintes situações para um conjunto de Dados:

    * conjunto com uma única moda (``UNIMODAL``),

    * com duas modas (``BIMODAL``),

    * com várias modas (``MULTIMODAL``) ou

    * nenhuma moda (``AMODAL``).


\\
Exemplo:     

```python
# Nossos Dados:

lista_1 = [11, 2, 2, 9, 7]
lista_2 = [4, 4, 19]
lista_3 = [3, 3, 1, 9, 9]


print("A moda de lista_1, é: ", sts.mode(lista_1))
print("A moda de lista_2, é: ", sts.mode(lista_2))
#print("A moda de lista_3, é: ", sts.multimode(lista_3)) <-- Aqui tive problemas de calcular para várias MODAS 

```

# Pesquisando sobre similaridade COSENO

* Link de estudo: https://stackoverflow.com/questions/18424228/cosine-similarity-between-2-number-lists

```python
from scipy import spatial

dataSetI = [3, 45, 7, 2]
dataSetII = [2, 54, 7, 15]
result = 1 - spatial.distance.cosine(dataSetI, dataSetII)

print(result)
```

```python
import math
def cosine_similarity(v1,v2):
    "compute cosine similarity of v1 to v2: (v1 dot v2)/{||v1||*||v2||)"
    sumxx, sumxy, sumyy = 0, 0, 0
    for i in range(len(v1)):
        x = v1[i]; y = v2[i]
        sumxx += x*x
        sumyy += y*y
        sumxy += x*y
    return sumxy/math.sqrt(sumxx*sumyy)

v1,v2 = [3, 45, 7, 2], [2, 54, 7, 15]
print(v1, v2, cosine_similarity(v1,v2))
```

```python
EPSILON = 1e-07 # Para garantir 

def cosine(x, y):
    dot_products = np.dot(x, y)
    norm_products = np.linalg.norm(x) * np.linalg.norm(y)
    return dot_products / (norm_products + EPSILON)
```

```python
v1,v2 = np.array([3, 45, 7, 2]), np.array([2, 54, 7, 15])

print(cosine(v1, v2))
```
